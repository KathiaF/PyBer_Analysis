```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create a Line Chart Using the Object-Oriented Interface
# Create the plot with ax.plt()
fig, ax = plt.subplots() #create a figure
ax.plot(x_axis, y_axis) #define the axes
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(), object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
#plt.show() should only be used once per session because it can cause the graph to display improperly. If you have problems with data not being displayed properly, you may have to restart your kernel and clear your output.
```

```python
#Annotate Charts Using the MATLAB Method
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()

# show the months January through June on the graph
plt.xlim(0, 5) #plt.xlim("Jan", "June")
#change the figure size to 20 pixels wide by 10 pixels high
plt.figure(figsize=(20,10)) #plt.subplots(figsize=(20,10))
```

```python
#Skill drill
#Using the object-oriented approach, make the following changes to your line chart:
# Create the plot.
fig, ax = plt.subplots() #create a figure
#define the axes
#Add a legend for the city of Boston.
ax.plot(x_axis, y_axis, marker="D", color="g", linewidth=2, label='Boston')
ax.legend()
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

#Add a title and axes labels.
ax.set_title("PyBer Fare by Months")

#Add a y-axis limit.
ax.set_ylim(0,50)

#Add grid lines.
ax.grid()
```

```python
# Create Scatter Plots and Bubble Charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis, color = "r", label = "Chicago")
plt.legend()
```

```python
# Create a Bubble Chart Using the MATLAB Method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#list comprehension
#new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
# SkillDrill
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], 
           label = 'Boston', color = 'skyblue', alpha = 0.8, 
           edgecolors = 'k', linewidths = 2)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")

ax.invert_yaxis()
```

```python
# Create Pie Charts
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize = (8,8)) #increase the size of the pie chart
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = '%.1f%%')
plt.show()
```
