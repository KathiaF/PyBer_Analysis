```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
# Make a reference to the books.csv file path
got_df = pd.read_csv("Resources/got.csv")
got_df.head()
```

```python
# Get attacker and defender king data
attacker_data = got_df["attacker_king"].value_counts()
defender_data = got_df["defender_king"].value_counts()
defender_data
```

```python
# Get total battle data
battle_data = attacker_data.add(defender_data, fill_value=0)
battle_data
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar", facecolor="r")

# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated in")
plt.xlabel("King")
plt.xticks(rotatio=45)
# Show plot
plt.show()

# Resize plot to display labels
plt.tight_layout()
```
